```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('merc.csv')
```

```python
df.head()
```

```python
df = df[df.year>2000]
plt.figure(figsize = (16, 8))
sns.scatterplot(data =df, x = 'mileage', y ='price', hue ='transmission', style ='fuelType', s=100)
# sns.regplot(data =df, x ='mileage', y='price')
plt.show()
```

1. `df = df[df.year>2000]` - This line filters the DataFrame `df` to include only the rows where the `year` column is greater than 2000. The result is reassigned to `df`, effectively removing any entries from the DataFrame that do not meet this condition.

2. `plt.figure(figsize = (16, 8))` - This line creates a new figure for plotting with a specified size of 16 inches in width and 8 inches in height.

3. `sns.scatterplot(data =df, x = 'mileage', y ='price', hue ='transmission', style ='fuelType', s=100)` - This line uses the Seaborn library to create a scatter plot. 
   - `data = df` specifies the DataFrame to use for the plot.
   - `x = 'mileage'` sets the x-axis to represent the `mileage` column.
   - `y = 'price'` sets the y-axis to represent the `price` column.
   - `hue = 'transmission'` colors the points based on the `transmission` column, allowing for differentiation between different transmission types.
   - `style = 'fuelType'` changes the marker style based on the `fuelType` column, providing another layer of differentiation.
   - `s=100` sets the size of the markers in the scatter plot to 100.

4. `plt.show()` - This line displays the figure with the scatter plot on the screen.

```python
sns.swarmplot(data =df, x =df['fuelType'], y = df['price'])
plt.show()
```


1. `sns.swarmplot(...)` - This function from the Seaborn library creates a swarm plot, which is a type of scatter plot that displays individual data points while avoiding overlap. It is particularly useful for visualizing categorical data.

2. `data=df` - This argument specifies the DataFrame (`df`) that contains the data to be plotted.

3. `x=df['fuelType']` - This sets the x-axis of the plot to the 'fuelType' column from the DataFrame, which represents different categories of fuel types.

4. `y=df['price']` - This sets the y-axis of the plot to the 'price' column from the DataFrame, representing the prices associated with each fuel type.

5. `plt.show()` - This function call displays the plot in the output. It is necessary to render the plot when using Matplotlib or Seaborn in a Jupyter Notebook.
